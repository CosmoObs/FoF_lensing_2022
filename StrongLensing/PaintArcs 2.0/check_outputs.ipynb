Image with [trilogy](https://github.com/oliveirara/trilogy):

```python
import trilogy
```

```python
for file in ['arceinstein', 'arc_conveinstein', 'arc_conv_noiseeinstein', \
             'backgroundeinstein', 'arc_conv_noise_bkgeinstein']:
    trilogy.Trilogy(infile=None, \
                    images=f'{file}'+'.fits', \
                    imagesorder='L', \
                    outname = f'{file}',
                    samplesize = 20000,
                    stampsize = 20000,
                    maxstampsize = 20000,
                    satpercent = 1e-3,
                    noiselum = 1e-1,
                    noisesig = 0,
                    noisesig0 = 0,
                    correctbias = 1,
                    colorsatfac = 0,
                    combine = "sum",
                    deletetests = 1,
                    deletefilters = 1,
                    testfirst = 0,
                    noiselums = {'L': 1e-1},
                    show = 1,
                    showwith = "PIL").run()
```
